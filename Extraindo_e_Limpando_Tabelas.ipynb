```python
import tabula
import pandas as pd
```

```python
dfs = tabula.read_pdf('files/dados.pdf', pages="all")
```

## Tratando dados da tabela de matérias

```python
tabela_materia = dfs[0]
tabela_materia
```

```python
tabela_materia.describe
```

```python
tabela_materia.drop('Unnamed: 0', axis=1, inplace=True)
tabela_materia.drop('Unnamed: 1', axis=1, inplace=True)
tabela_materia.columns.name = 'id'
tabela_materia.drop(0, inplace=True)
tabela_materia
```

```python
tabela_materia.rename(columns={'Tabela 1 - Matéria': 'descricao', 'Unnamed: 2':'nota_minima'}, inplace=True)
```

```python
tabela_materia # Tabela materia pronta
```

```python
tabela_materia.to_excel('files/tabela_materia.xlsx') # Exportando tabela_materia
```

## Tratando dados da tabela de estudantes

```python
tabela_estudante = dfs[4]
tabela_estudante.drop('Unnamed: 0', axis=1, inplace=True)
tabela_estudante.drop('id', axis=1, inplace=True)
tabela_estudante.columns.name = 'id'
tabela_estudante
```

```python
tabela_estudante.to_excel('files/tabela_estudante.xlsx')
```

## Tratando dados da tabela de notas (parte 1)

```python
dfs[5]
```

```python
tabela_notas = dfs[5].set_index('data_inicio_trimestre')
tabela_notas.drop('Unnamed: 0', axis=1, inplace=True)
tabela_notas.drop('Unnamed: 1', axis=1, inplace=True)
tabela_notas.drop('nota', axis=1, inplace=True)
tabela_notas.drop('faltas', axis=1, inplace=True)
tabela_notas
```

```python
tabela_notas.rename(columns={'Unnamed: 2': 'nota', 'Unnamed: 3': 'faltas'}, inplace=True)
tabela_notas
```

```python
tabela_notas.to_excel('files/files_aux/tabela_nota_aux.xlsx')
```
